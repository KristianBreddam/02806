```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
# Standard plotly imports
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
```

```python
df = pd.read_csv("gadetraer.csv")
df.head()
```

```python
df.shape
```

```python
traer = df.groupby(['slaegtsnavn']).count().id
traer = traer.sort_values(ascending=False)
```

```python
plt.rcParams['figure.figsize'] = (12, 5)
plt.bar(traer.index,traer.values)
plt.title("SlÃ¦gtsnavn")
plt.ylabel("Tree counts")
plt.xticks(rotation=90)
plt.show()
```

```python
byer = df.groupby(['bydelsnavn']).count().id
byer = byer.sort_values(ascending=False)
```

```python
plt.rcParams['figure.figsize'] = (12, 5)
plt.bar(byer.index,byer.values)
plt.title("Bydele")
plt.ylabel("Tree counts")
plt.xticks(rotation=45)
plt.show()
```
