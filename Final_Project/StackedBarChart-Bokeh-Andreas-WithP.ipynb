```python
temp = df.groupby(['bydelsnavn','slaegtsnavn_eng']).count().id.unstack()
temp = temp.fillna(0) # some city areas do not have certain trees
temp = temp.reset_index()
temp
```

```python
# Add the percentages 
for tree in temp.columns[1:]:
    temp2 = []
    for values in temp[tree]:
        temp2.append(round(values/sum(temp[tree])*100,1))
        colName = "{}_p".format(tree)
    temp[colName] = temp2
```
